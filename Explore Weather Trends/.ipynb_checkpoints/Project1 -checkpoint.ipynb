# Exploring Weather Trends, DAND Project 1:

##### In This Project, I'm going to use a Data about Alexandria city in Egypt, Contains yearly-average temperature of the City, Against the global temperature.

Data is extracted from two tables in a PostgreSql Database using sql-queries. <br>
1- city_data - This contains the average temperatures for each city by year (ºC). <br>
        
        SELECT * 
        FROM city_data
        WHERE city = 'Alexandria' AND country = 'Egypt'; 
        
    
2- global_data - This contains the average global temperatures by year (ºC). <br>

        SELECT *
        FROM global_data;


--------------------------

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np

plt.rcParams["figure.figsize"] = [10, 8]
```

### IMPORTING DATA:

```python
city_data     = pd.read_csv('city_data.csv')

global_data = pd.read_csv('global_data.csv')
```

```python
city_data.head()
```

```python
global_data.head()
```

##### Moving Average for the Global data and City data.
This is used to smooth out data and to make it easier to observe long term trends. <br> 

Rolling Average has been calculated to smooth out data and to make it easier to observe the trends when it be shown in Charts. It's been calculated for every 10 years to each single data but the first 10.

```python
global_avg = global_data['avg_temp'].rolling(12).mean()

city_avg = city_data['avg_temp'].rolling(12).mean()
```

### VISUALISATIONS:

```python
plt.plot(global_data['year'], global_avg, label='Global')
plt.plot(city_data['year'], city_avg, label='Alex')

plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature") 
plt.title("Alexandria Average Temperature")
plt.show()
```

```python
plt.plot(global_data['year'], global_avg, label='Global')

plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature") 
plt.title("Global AVG Temperature")
plt.show();
```

```python
plt.plot(city_data['year'], city_avg, label='Alex')

plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature") 
plt.title("Alexandria Average Temperature")
plt.show();
```
