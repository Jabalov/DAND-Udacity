### IMPORTING PYTHON DATA SCIENCE PACKAGES, USED IN THE ANALYSIS OF DATA.

```python
import pandas as pd 

import matplotlib.pyplot as plt 
import numpy as np

plt.rcParams["figure.figsize"] = [10, 8]
```

## OPENING UP THE CSV FILES

```python
city_data     = pd.read_csv('city_data.csv')

global_data = pd.read_csv('global_data.csv')
```

### Moving Average for the Global data and City data.

Now that I could extract data from a database, let's talk through the next step of completing the project: moving averages. <br>
Moving averages are used to smooth out data to make it easier to observe long term trends and not get lost in daily fluctuations <br> 
Using a moving average, you can both smooth out the daily volatility and allow you to observe the long term trend.


### Moving Average with Python:

Simple, cumulative, and exponential moving averages with Pandas <br>
The moving average is also known as rolling mean. <br>

The easiest way to calculate the simple moving average is by using the <code> pandas.Series.rolling </code> method.

```python
global_avg = global_data['avg_temp'].rolling(20, min_periods=2).mean()

city_avg = city_data['avg_temp'].rolling(20, min_periods=2).mean()
```

--------------------


### CREATING THE LINE CHARTS

#### Visualisations is done by matplotlib.pyplot python library.

```python
plt.plot(global_data['year'], global_avg, label='Global')
plt.plot(city_data['year'], city_avg, label='Alex')

plt.legend()
plt.xlabel("Years", fontsize=16)
plt.ylabel("Temperature", fontsize=16) 
plt.title("Local Average VS. Global Average Temperature", fontsize=16)
plt.show()
```

```python
plt.plot(global_data['year'], global_avg, label='Global')

plt.legend()
plt.xlabel("Years",  fontsize=16)
plt.ylabel("Temperature",  fontsize=16) 
plt.title("Global AVG Temperature",  fontsize=16)
plt.show();
```

```python
plt.plot(city_data['year'], city_avg, label='Alex')

plt.legend()
plt.xlabel("Years",  fontsize=16)
plt.ylabel("Temperature",  fontsize=16) 
plt.title("Alexandria AVG Temperature", fontsize=16)
plt.show();
```
